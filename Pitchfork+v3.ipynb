```python
# create bin dividers

bins = [0,3,6,8,10]

# bin names
group_names = ['awful','ok','fine','great']

df_simple['score_desc'] = pd.cut(df_simple['score'], bins, labels=group_names)
```

```python
df_genre = pd.get_dummies(df_simple['genre'])

df_simple = pd.concat([df_simple, df_genre], axis=1)
df_simple
```
